```python
import numpy as np
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
import matplotlib.pyplot as plt
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
import numpy as np
from scipy import linalg

# Create the 2x2 matrix
matrix = np.array([[4, 5],
                   [3, 2]])

# Calculate the determinant
det = linalg.det(matrix)

# Print the result
print("Determinant:", det)
```

```python
import numpy as np
from scipy import linalg

# Original matrix
matrix = np.array([[4, 5],
                   [3, 2]])

# Calculate inverse
inverse_matrix = linalg.inv(matrix)

# Print the inverse
print("Inverse of the matrix:")
print(inverse_matrix)
```

```python
import numpy as np
from scipy.sparse import csr_matrix, csc_matrix

# Define data for matrix A
data_A = np.array([1, 2, 3])
row_A = np.array([0, 1, 2])
col_A = np.array([0, 1, 2])
A = csr_matrix((data_A, (row_A, col_A)), shape=(3, 3))

# Define data for matrix B
data_B = np.array([4, 5, 6])
row_B = np.array([0, 1, 2])
col_B = np.array([2, 1, 0])
B = csc_matrix((data_B, (row_B, col_B)), shape=(3, 3))
# Convert to dense array
print("Matrix A (dense):\n", A.toarray())

# Count non-zero elements
print("Non-zero count in A:", A.count_nonzero())

# Transpose matrix B
print("Matrix B Transposed:\n", B.transpose().toarray())

# Matrix multiplication
C = A @ B
print("A @ B:\n", C.toarray())
```

```python
import numpy as np
from scipy import linalg

matrix = np.array([[5, 4],
                   [6, 3]])

eigenvalues, eigenvectors = linalg.eig(matrix)

print("Eigenvalues:")
print(eigenvalues)

print("Eigenvectors:")
print(eigenvectors)
```
